```python
import random as r
import matplotlib.pyplot as plt
from captcha.image import ImageCaptcha

num = "1234567890"
small_key="abcdefghijklmnopqrstuvwxyz"
cap_key="ABCDEFGHIJKLMNOPQRSTUVWXYZ"

cap_char = cap_key+small_key+num
#instantiationg ImageCaptcha by calling constructor automatic object instantiation 
generator = ImageCaptcha()

random_str = ''.join([r.choice(cap_char) for i in range(4)])

print(random_str)

img = generator.generate_image(random_str)

plt.imshow(img)
#plt.title(random_str)
plt.show()

#saving verification code

fname = 'images/'+random_str+'.jpg'
#img.save(fname)

img.show()

```
